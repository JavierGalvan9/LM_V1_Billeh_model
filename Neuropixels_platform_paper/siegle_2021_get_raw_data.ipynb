```python
from pynwb import NWBHDF5IO

nwb_file_path = '/home/jgalvan/Desktop/Neurocoding/LM_V1_Billeh_model/Neuropixels_data/sub-703279277_ses-719161530.nwb'
# Open the .nwb file
with NWBHDF5IO(nwb_file_path, 'r') as io:
    # Read the NWB file
    nwb_file = io.read()
    gabor_data = nwb_file.intervals['gabors_presentations']
    # iterate over columns of gabor data
    gabor_dict = {}
    for key, data in zip(gabor_data.colnames, gabor_data.columns):
        print(key, data[:])
        gabor_dict[key] = data[:]
```
